```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os, sys, shutil
```

```python
workspace = '/data/ChaochaoData/DSA-DL/HyperMorph/Comparison/'
sample_dir = os.path.join(workspace, 'likert_samples')
eval_dir = os.path.join(workspace, 'likert_scoring')
```

```python
# score_filename = 'LikertScoring_DC.xlsx'
score_filename = 'LikertScoring_SA.xlsx'
# score_filename = 'LikertScoring_RA.xlsx'

filepath = os.path.join(eval_dir, score_filename)
df = pd.read_excel(filepath)
print(df.shape)
df
```

```python
## Read right series labels
def readRightLabels(score_filename):
    filepath = os.path.join(eval_dir, score_filename)
    df = pd.read_excel(filepath)
    case_nums = df['Case']

    N_case = len(df)
    right_labels = []
    for i in range(N_case):
        dst_dir = os.path.join(sample_dir, str(case_nums[i]).zfill(3))
        with open(os.path.join(dst_dir, "pool_order.txt"), "r") as file:
            pool_order = [int(line) for line in file.readlines()]
        right_labels.append(pool_order[1])

    right_labels = np.array(right_labels) 
    return right_labels
```

```python
right_labels = readRightLabels(score_filename)
df['Right Labels'] = right_labels
df
```

```python
measure_name = 'Vascular Fidelity'
# measure_name = 'Subtraction Artifact'
# measure_name = 'Overall Quality'

measure_scores = df[measure_name]
dsa_scores = measure_scores[right_labels == 0]
reg_scores = measure_scores[right_labels == 1]

print(f"{'Score':5}; {'DSA':5}; {'Reg':5}")
for i in range(5,0,-1):
    print(f"{i:5}; {np.sum(dsa_scores==i):5}; {np.sum(reg_scores==i):5}")
print(f"{'Total':5}; {len(dsa_scores):5}; {len(reg_scores):5}")
print(f"{'Avg':5}; {np.mean(dsa_scores):5.1f}; {np.mean(reg_scores):5.1f}")
```
