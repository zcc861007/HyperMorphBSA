```python
## Read left series labels
def readLeftLabels(score_filename):
    filepath = os.path.join(eval_dir, score_filename)
    df = pd.read_excel(filepath)
    case_nums = df['Case Number']

    N_case = len(df)
    left_labels = []
    for i in range(N_case):
        dst_dir = os.path.join(sample_dir, str(case_nums[i]).zfill(3))
        with open(os.path.join(dst_dir, "pool_order.txt"), "r") as file:
            pool_order = [int(line) for line in file.readlines()]
        left_labels.append(pool_order[0])

    left_labels = np.array(left_labels)
    return left_labels
```

```python
left_labels = readLeftLabels(score_filename)
df['Left Labels'] = left_labels
df[:10]
```

```python
measure_name = 'Pixel Shift vs Voxel Morph Grading'

measure_scores = df[measure_name]
dsa_scores = 6-measure_scores[left_labels == 0]
reg_scores = 6-measure_scores[left_labels == 1]

print(f"{'Score':5}; {'DSA':5}; {'Reg':5}")
for i in range(5,0,-1):
    print(f"{i:5}; {np.sum(dsa_scores==i):5}; {np.sum(reg_scores==i):5}")
print(f"{'Total':5}; {len(dsa_scores):5}; {len(reg_scores):5}")
print(f"{'Avg':5}; {np.mean(dsa_scores):5.1f}; {np.mean(reg_scores):5.1f}")
```
